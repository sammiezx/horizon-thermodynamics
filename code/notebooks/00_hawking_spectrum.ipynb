# Hawking spectrum — a first look

A simple starter notebook: plot the thermal Hawking spectrum for a few black-hole
masses to get a feel for the temperature scale and the Wien cutoff.

This is the *thermal* spectrum — greybody factors not yet included.

```python
import numpy as np
import matplotlib.pyplot as plt

from bh_radiation.hawking import hawking_temperature, planck_spectrum, HBAR, K_B
```

```python
# Solar-mass BH (cold), 10^12 kg primordial BH (hot)
M_sun = 1.989e30
M_primordial = 1e12

for M, label in [(M_sun, '1 M_sun'), (M_primordial, '1e12 kg')]:
    T = hawking_temperature(M)
    print(f'{label:>12}:  T_H = {T:.3e} K   (kT = {K_B*T:.3e} J)')
```

```python
# Plot in natural units to see the universal shape.
x = np.linspace(0.01, 15, 500)   # x = hbar*omega / kT
u = planck_spectrum(x, temperature=1.0, natural_units=True)

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.plot(x, u, lw=2)
ax.set_xlabel(r'$\hbar\omega / k_B T_H$')
ax.set_ylabel(r'spectral energy density (natural units)')
ax.set_title('Thermal Hawking spectrum (no greybody factor)')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
